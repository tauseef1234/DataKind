## Dataframe 
Create a single output dataframe as per the requirements for the data from year 2017-2022.



```python
# create a empty dataframe which will hold the final data
df_final = pd.DataFrame()
for root, _, files in os.walk("/content/drive/My Drive/Raw Census and Capacity Numbers"):
    # print(root)
    for file_name in files:
      # read only xlsx files, files that start with a date and excluding 2016 files
      if (file_name.partition('.')[2] == 'xlsx') & (re.match('[0-9]', file_name.partition('.')[0]) is not None)& (file_name.split('.')[0].split('-')[-1]!='2016'):
        df = pd.DataFrame()
        final_df = pd.DataFrame()
        print (os.path.join(root, file_name))

        # read the excel file
        df = pd.read_excel(os.path.join(root, file_name))
        # filter out the redundant blank rows before the data
        df = df[df[df[df.columns[0]]=='MEN'].index[0]:].copy()
        # rename columns with the first row
        df.rename(columns=df.iloc[0],inplace=True)
        # exclude rows with the header name
        df=df[1:].copy()
        df.reset_index(drop=True,inplace=True)

        # rename columns
        df.columns = [df.iloc[0][i] if pd.isna(col) else col for i, col in enumerate(df.columns)]
        # check for files with old/new formatting and rename them acordingly
        if [col for col in df.columns if str(col).find('Current') !=-1]:
          df.rename(columns={df.columns[0]:'Provider Name',
                             df.columns[2]:'Ward',
                             'Current Bed Availability ':'Bed Availability ',
                             list(filter(lambda x: 'Served' in str(x), df.columns))[0]:'Clients Served'},inplace = True)
        else:
          df.rename(columns={df.columns[0]:'Provider Name',
                             df.columns[2]:'Ward',
                             list(filter(lambda x: 'Served' in str(x), df.columns))[0]:'Clients Served'},inplace = True)
        
        # keep only the relevant columns 
        df=df[['Provider Name','Phone Number','Ward','Bed Availability ','Clients Served','Vacancies']].copy()
        if df[df['Provider Name']=='FAMILIES'].empty:
          indx = df[df.isnull().all(axis=1)].index[-3]
        else:
          indx = df[df['Provider Name']=='FAMILIES'].index[0]
        # separate women and men data
        women_df = df[df[df['Provider Name']=='WOMEN'].index[0]+2:indx].copy()
        women_df['Gender Served'] ='Female'
        men_df = df[:df[df['Provider Name']=='WOMEN'].index[0]].copy()
        men_df['Gender Served'] = 'Male'
        # append both men and women dataframes
        final_df = men_df.append(women_df).reset_index(drop=True)
        final_df['Date'] = file_name.partition('.')[0]
        final_df = final_df[~final_df['Provider Name'].isna()].copy()
        final_df.reset_index(drop=True,inplace=True)
        # append the dataframe for a single day in one dataframe
        df_final = df_final.append(final_df)
```

```python
df_final['Provider Name'] = df_final['Provider Name'].str.replace('\n',' ')
```

```python
df_final['Month'] = pd.to_datetime(df_final['Date']).dt.month
df_final['Year']= pd.to_datetime(df_final['Date']).dt.year
df_final['Month_Year']= df_final['Month'].astype(str) +'-'+ df_final['Year'].astype(str)
```
